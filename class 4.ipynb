```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None
        
        
    def append(self, data):
        current = self.head
        if self.head == None:
            self.head = Node(data)
        else: 
            while current.next:
                current = current.next 
            current.next = Node(data)
            
            
    def insert(self, position_after, target_value):
        current = self.head
        counter = 0
        if not self.head:
            self.head = Node(target_value)
        else:
            while current.next and position_after != counter:
                current = current.next
                counter += 1 
        # step 1:
                old_next = current.next      # [1, 2, 3, 4, 5]
        # step 2:                   # new_node = N(5, None)
                new_node = Node(target_value) # current(3, None); old_node = None
                current.next = new_node        # current(3, new_node)
        # step 3:
                new_node.next = old_next      # new_node(5, old_next)
        
        
        
    
    def print_list(self):
        out_str = "["
        current = self.head 
        if self.head:
            out_str += "%s" % self.head.data
            current = self.head.next 
        while current:
            out_str += ", %s" % current.data
            current = current.next 
        out_str += "]"
        return out_str
        
    def __str__(self):
        return self.print_list()
        
    def index(self, target):
        current = self.head 
        counter = 0
        while current:
            if current.data == target:
                return counter 
            current = current.next
            counter += 1
        raise ValueError("%s is not in the list" % target)
        
       
    def get_value_at_index(self, index):
        current = self.head
        counter = 0
        while current:
            if current == index:
                return current.data
            current = current.next
            counter += 1
        raise IndexError("list index out of range")

    def __getitem__(self, item):
        return self.get_value_at_index(item)
            
            
    def remove(self, target_value):
        current = self.head
        found = False 
        if current.data == target_value:
            self.head = self.head.next
        else:
            while current.next and current.data != target_value:
                prev = current                 # prev = 1; prev =2
                current = current.next         # current = 2; current =3 
        if current.data == target_value:       # 2 != 3; 3 == 3
                found = True                   # found = True
                prev.next = current.next       # 2 -> 4:
        if not found: 
            raise ValueError("%s is not in list")
        
        
        
        
            
            
        
        
        
        
```
